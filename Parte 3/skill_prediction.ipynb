# Skill prediction en Linkedin

#### a. Generación de la matriz dispersa Z a partir de los datos que se encuentran en el archivo *user_skill*. La matriz genera es una matriz dispersa comprimida por filas.

```python
import re
from scipy.sparse import csr_matrix
import numpy as np

def parse_user_skills():
    user_skills = {}
    with open("data/user_skill") as f:
        for line in f:
            (user_id, skills) = line.split(":")
            user_skills[int(user_id)] = [int(skill) for skill in skills.split(",")]
    
    n = len(user_skills)  # number of users
    print "Número de usuarios: {}".format(n)
    d = max(i for v in user_skills.itervalues() for i in v) + 1  # number of skills
    print "Número de skills: {}".format(d)
    
    row = []
    col = []
    data = []
    # data, row_ind and col_ind satisfy the relationship a[row_ind[k], col_ind[k]] = data[k].
    for user, skills in user_skills.iteritems():
        for skill in skills:
            row.append(user)
            col.append(skill)
            data.append(1)
    
    Z = csr_matrix((data, (row, col)), shape=(n, d))  
    return Z

Z = parse_user_skills()
```

#### b. Construcción de las sub-matrices Z_train y Z_test, las cuales representan los conjuntos de entrenamiento y de pruebas respectivamente. 
Para lo anterior se utiliza la función *train_test_split* del módulo model_selection de sklearn, indicando el tamaño de del conjunto de entrenamiento y una semilla para realizar selección aleatoria de las filas. Las submatrices generadas son conjuntos disjuntos entre sí y subconjuntos de filas de la matriz original Z.

```python
from sklearn.model_selection import train_test_split
Z_train, Z_test = train_test_split(Z, train_size=0.7, random_state=42)
print Z_train.shape
print Z_test.shape
```

#### d. A partir de las matrices Z_train y Z_test, se generan las matrices de variables predictores X y de etiquetas y, considerando la elección de una skill.
Se elije la columna j (id de la skill escogida) de cada matriz (Z_train y Z_test). Esta columna es el nuevo vector de etiquetas. La skill escogida es la de id = 73 (Microsoft Word).

```python
def get_X_y_matrices(Z, skill):
    n, d = Z.shape
    Z_col = Z.tocsc(copy=True)  # csc matrices are fast in column slicing
    X = Z_col[:,:-1]
    if skill < d - 1:
        X[:,skill:] = Z_col[:,skill+1:]

        
    y = Z.getcol(skill).toarray().ravel()
    X = X.tocsr(copy=True)
    
    return X, y

skill_id = 73  # Microsoft Word
    
X_train, y_train = get_X_y_matrices(Z_train, skill_id)
X_test, y_test = get_X_y_matrices(Z_test, skill_id)
```

#### e. Prueba de cinco clasificadores distintos. Los elegidos son: SVM con un kernel lineal, Multinomial Naive Bayes, Regresión Logística, K Vecinos más cercanos y árbol de decisión.
Las métricas utilizadas para evaluar los clasificadores son *accuracy*, *precision*, *recalll* y la matriz de confusión. Se saca provecho de que el problema de clasificación es binario (está o no está presente una skill). **Estas métricas estarán basadas en el rendimiento obtenido en clasificar el dataset de pruebas.** A continuación, se presentará una breve definición de las métricas anteriormente mencionadas:

* **Matriz de confusión:**
Por definición, una matriz de confusión $C$ es tal que $C_{i,j}$ es igual al número de observaciones que se sabe que pertenecen al grupo $i$ pero que se predicen como parte del grupo $j$.
Así que, en clasificación binaria, $C_{0,0}$ es la cantidad de negativos verdaderos (True Negatives (TN)), $C_{1,0}$ son los falsos negativos (False Negatives (FN)), $C_{1,1}$ son los verdaderos positivos (True Positives (TP)), y $C_{0,1}$ son los falsos positivos (False Positives (FP)).

* **Accuracy**: 
$$Accuracy = \frac{TP+TN}{TP+TN+FP+FN} = \frac{correct\,predictions}{all\,predictions}$$

* **Precision**:
La precisión cuantifica la habilidad del clasificador de no etiquetar un ejemplo negativo como positivo. Se puede interpretar como la probabilidad de que una predicción positiva hecha por el clasificador sea positiva. Matemáticamente se define como:

$$Precision = \frac{TP}{TP+FP}$$

* **Recall**:
Es el número de instancias correctamente clasificadas como positivo (true positives) dividido por el número total de positivos en el conjunto de prueba (true positives + false negatives). Matemáticamente se define como: 

$$Recall = \frac{TP}{TP+FN}$$

```python
from sklearn.tree import DecisionTreeClassifier as DTC
from sklearn.svm import SVC
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier as KNC
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import BernoulliNB as BNB
from sklearn.naive_bayes import MultinomialNB as MNB
from sklearn.linear_model import LogisticRegression as LReg
from sklearn.metrics import (accuracy_score, precision_score, recall_score, confusion_matrix)
import pandas as pd
from IPython.display import HTML
import itertools

clfs = {
    "SVC with linear kernel": LinearSVC(),
    "Multinomial Naive Bayes": MNB(),
    "Logistic Regression": LReg(),
    "K Nearest Neighbors": KNC(),
    "Decision Tree": DTC(max_depth=5)
}

prediction = {}
for model_name, model in clfs.iteritems():
    model.fit(X_train, y_train)
    prediction[model_name] = model.predict(X_test)

scores = {"model":[], "accuracy":[], "precision":[], "recall":[]}
confusion_matrices = {}
for model_name, y_pred in prediction.iteritems():
    scores["model"].append(model_name)
    scores["accuracy"].append(accuracy_score(y_test, y_pred))
    scores["precision"].append(precision_score(y_test, y_pred))
    scores["recall"].append(recall_score(y_test, y_pred))
    confusion_matrices[model_name] = confusion_matrix(y_test, y_pred)


scores_df = pd.DataFrame.from_dict(scores)
scores_df = scores_df[["model", "accuracy", "precision", "recall"]]
scores_df.set_index("model", inplace=True)
HTML(scores_df.to_html())
```

#### f. Misma prueba que en el item anterior, pero ahora variando el tamaño porcentual del dataset de entrenamiento (10%-90%) y cambiando las skills analizadas. De estas últimas, se escogieron de tal forma que tuvieran un número distinto de usuarios para ver el efecto del balance en las clases.

```python
skill_ids = [21, 30, 29]  # Process Improvement (113), Mobile Technology (31), Microsoft Office (3081)
train_sizes = np.linspace(0.1, 0.9, 9)
data = {id: {model_name:{"accuracy":[], "precision":[], "recall":[]} for model_name in clfs.keys()} for id in skill_ids}
for skill_id in skill_ids:
    for size in train_sizes:
        Z_train, Z_test = train_test_split(Z, train_size=size)
        X_train, y_train = get_X_y_matrices(Z_train, skill_id)
        X_test, y_test = get_X_y_matrices(Z_test, skill_id)
        for model_name, model in clfs.iteritems():        
            model.fit(X_train, y_train)
            y_pred = model.predict(X_test)
            data[skill_id][model_name]["accuracy"].append((size, accuracy_score(y_test, y_pred)))
            data[skill_id][model_name]["recall"].append((size, recall_score(y_test, y_pred)))
            data[skill_id][model_name]["precision"].append((size, precision_score(y_test, y_pred)))


for score_name in ["accuracy", "precision", "recall"]:
    fig, axs = plt.subplots(ncols=3, figsize=(15, 6))
    fig.tight_layout()
    for i, skill_id in enumerate(skill_ids):        
        for model_name in clfs.keys():
            axs[i].plot(*zip(*data[skill_id][model_name][score_name]), label=model_name)

        axs[i].set_xlabel('Training set proportion')
        axs[i].set_xlim([min(train_sizes), max(train_sizes)])
        axs[i].set_ylabel(score_name)
        axs[i].legend(loc="lower left", frameon=False, prop={'size':10})
        axs[i].set_title(skills[skill_id], size=20)
```
